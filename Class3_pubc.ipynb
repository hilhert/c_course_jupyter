####  定长二进制数
计算机内部2进制数的表达受到`位数`的限制<br>
例： 8位二进制数对于`整数`的表达范围为$-128 \to 127$(整数分为正数与负数)

8位二进制数对于`正数`的表达范围为$0 \to 255$

range(<u>&nbsp;&nbsp;</u>&nbsp;&nbsp;<u>&nbsp;&nbsp;</u>&nbsp;&nbsp;<u>&nbsp;&nbsp;</u>&nbsp;&nbsp;<u>&nbsp;&nbsp;</u>&nbsp;&nbsp;<u>&nbsp;&nbsp;</u>&nbsp;&nbsp;<u>&nbsp;&nbsp;</u>&nbsp;&nbsp;<u>&nbsp;&nbsp;</u>&nbsp;&nbsp;<u>&nbsp;&nbsp;</u>&nbsp;&nbsp;)=$2 \times 2 \times 2 \times \cdots$=$2^8$＝256

从0数到255；从－128数到127统统都是256

代码示例：
```C
signed char a=127,b=3,c;
c=a+b;
printf("%d",c);
```

正整数的最高位（定长才成立）为0<br>
负整数的最高位（定长才成立）为1<br>
**问：8位二进制数1000&nbsp;&nbsp;0011代表-3吗？**<br>

````c
#include <stdio.h>
int main(){
printf( "%d\n" ,(signed char)0x83);
return 0;
}
````

如上，答案是否定的。Why？（注意数字0）<br>
因为：计算机内部`**负**整数`采用`补码`存放<br>
**补码的生成： 将（-D）去号，转换为2进制数，然后各位取反，再加一** <br>
0000&nbsp;&nbsp;0011 $\rightarrow$1111&nbsp;&nbsp;1100(反码)$\rightarrow$1111&nbsp;&nbsp;1101$\rightarrow$0xFD

````c
#include <stdio.h>
int main(){
printf( "%d\n" ,(signed char)0xFD);
return 0;
}
````

#溢出归纳
1. 有符号整型发生上溢（运算结果超出类型表达的最大整数）有可能转换为补码形式的负数，此时最高位为1<br>
2. 有符号整型发生下溢（运算结果超出类型表达的最小整数）有可能转换为原码形式的整数，此时最高为为0<br>
3. 无符号整型发生上溢（超出数值表达范围）即转换为模K运算，K为`位限数`（如：8位二进制数的`位限数`为：256）<br>
4. 无符号整型发生下溢（运算结果小于0）将减法转换为减数的补码，再作加法<br>
**发生溢出的运算通通视为无效运算,无效运算在C语言中比比皆是,随处可见！**<br>

````c
//堆排序
#include <stdio.h>
void swap(int* a, int*b){
int tmp = *a;
*a      = *b;
*b      = tmp;
}
void adjust_heap(int* arr, int top,int N){
int tmp = arr[top];
for(int j=top*2+1;j<N;j=j*2+1){
if(arr[j]<arr[j+1]&&(j+1)<N) j++;
if(tmp<arr[j]){ 
arr[top]=arr[j];
top = j;
}else break;
}
arr[top] = tmp;

}
void create_heap(int* arr, int N){
for(int i=N/2-1;i>=0;i--) adjust_heap(arr,i,N);
}
void print_n(int* arr, int n){
for(int i=0;i<n;i++) printf("%d  ",arr[i]);
printf("\n");
}
int main(){
int arr[17]={45,23,12,-15,76,34,77,23,-66,34,89,20,33,56,79,-21,-3};
create_heap(arr,17);
//printf("%d",arr[0]);

for(int n=17;n>1;n--)
{
swap(&arr[0],&arr[n-1]);
adjust_heap(arr,0,n-1);
print_n(arr,17);
}

return 0;
}

````

````c
//快速排序
void swap(int* a, int*b){
int tmp = *a;
*a      = *b;
*b      = tmp;
}
#include <stdio.h>
void print_n(int* arr, int n){
for(int i=0;i<n;i++) printf("%d  ",arr[i]);
printf("\n");
}
int put_last(int*arr, int left ,int right){
int i=left,j=right,tmp=arr[i];
while(i<j){
while(i<j&&arr[j]>=tmp){j--;}; arr[i]=arr[j];
while(i<j&&arr[i]<=tmp){i++;}; arr[j]=arr[i];
}
arr[i]=tmp;
return i;
}
void quick_sort(int* arr, int left,int right){
if(left<right){
int b = put_last(arr,left,right);
quick_sort(arr,left,b-1);
quick_sort(arr,b+1,right);
}
}

int main(){
int arr[17]={45,23,12,-15,76,34,77,23,-66,34,89,20,33,56,79,-21,-3};
quick_sort(arr,0,16);
print_n(arr,17);
return 0;
}
````
